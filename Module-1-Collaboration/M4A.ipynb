```python
import sqlalchemy as sqla
```

```python
conn = sqla.create_engine('sqlite:///books.db')
```

```python
with conn.connect() as connection:
    query = sqla.text('SELECT title FROM books ORDER BY title ASC')
    rows = connection.execute(query)
    for row in rows:
        print(row[0])
```
