```python
class Solution:
    def sort012(self, arr):
        sorted_arr = []
        for x in range(3):  
            for n in arr:
                if n == x:
                    sorted_arr.append(n)
        arr[:] = sorted_arr
```

```python
class Solution:
    def binarysearch(self, arr, k):
        low = 0
        high = len(arr) - 1
        result = -1
        
        while low <= high:
            mid = (low + high) // 2
            
            if arr[mid] == k:
                result = mid
                high = mid - 1
            elif arr[mid] < k:
                low = mid + 1
            else:
                high = mid - 1
        
        return result
```
