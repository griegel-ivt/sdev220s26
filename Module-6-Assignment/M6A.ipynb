```python
from multiprocessing.dummy import Process
import random
import time
from datetime import datetime

def process_task():
    time.sleep(random.random())
    current_time = datetime.now()
    print(current_time)

if __name__ == "__main__":
    for i in range(3):
        p = Process(target=process_task)
        p.start()
```
